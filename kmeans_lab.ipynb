```python
import pandas as pd
```

```python
top_100_total=pd.read_csv('top_100_total.csv')
not_hot_total=pd.read_csv('not_hot_total.csv')
```

```python
top_100_total['song_type'] = 'hot'
```

```python
not_hot_total['song_type'] = 'not'
```

```python
data = pd.concat([top_100_total, not_hot_total], axis=0)
```

```python
data.head()
```

```python
data = data.reset_index(drop = True)
```

```python
display(data)
```

```python
data.describe()
```

```python
from sklearn import datasets # sklearn comes with some toy datasets to practise
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
numb= data[["energy", "speechiness","acousticness", "instrumentalness", "loudness","tempo","danceability","valence","liveness", "time_signature", "key"]]
```

```python
numb.head()
```

```python
X=numb
```

```python
import pickle

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X) #this is numpy array, we transform it below
filename = "scaler.pickle" # Path with filename #first creat an empty file and then save the scaler in it

with open(filename, "wb") as file: #this approach makes the file uncorruptable when creating it becuase it closes it before it finishes the process
        pickle.dump(scaler,file)

X_scaled_df = pd.DataFrame(X_scaled, columns = X.columns)
display(X.head())
print()
display(X_scaled_df.head())
```

```python
kmeans = KMeans(n_clusters=15, random_state=1234)
kmeans.fit(X_scaled_df)
```

```python
K = range(2, 21)
inertia = []
silhouette = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                    verbose=1) #displays a progress message
    kmeans.fit(X_scaled_df)
    filename = "kmeans_" + str(k) + ".pickle" # Path with filename
    with open(filename, "wb") as file:
        pickle.dump(kmeans,file)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X_scaled_df, kmeans.predict(X_scaled_df)))


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(K, inertia, 'bx-')
ax[0].set_xlabel('k')
ax[0].set_ylabel('inertia')
ax[0].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[0].set_title('Elbow Method showing the optimal k')
ax[1].plot(K, silhouette, 'bx-')
ax[1].set_xlabel('k')
ax[1].set_ylabel('silhouette score')
ax[1].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[1].set_title('Silhouette Method showing the optimal k')
```
